```
!pip install ibm_watson==6.1.0
from ibm_watson import LanguageTranslatorV3
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

```

apikey_lt=IAMAuthenticator("vg7kPENua9nsny6UJTAPBJO1bIEZIbiGmNYaKDYLNBPf")
version_lt="2018-05-01"
language_translator=LanguageTranslatorV3(authenticator=apikey_lt, version=version_lt)
language_translator.set_service_url('https://api.au-syd.language-translator.watson.cloud.ibm.com/instances/b405a34b-b855-419f-8b2b-04d8c62d85a4')
original_text=input("Text:")
english_to_spanish=language_translator.translate(text=original_text, model_id="en-es").get_result()
english_translation=english_to_spanish["translations"][0]["translation"]
print("Texto traducido español: "+ english_translation)

spanish_to_french=language_translator.translate(text=english_translation, model_id="es-fr"). get_result()
french_translation=spanish_to_french["translations"][0]["translation"]
print("Texto traducido frances: " + french_translation)





```
